```python
from ragatouille import RAGPretrainedModel

RAG = RAGPretrainedModel.from_pretrained("antoinelouis/colbertv1-camembert-base-mmarcoFR", n_gpu=1)
```

```python
from ragatouille.data import CorpusProcessor, llama_index_sentence_splitter
import os

path = '/home/adrien/Documents/Coding/RAG_ASN/Ingestion/ASN/lettres_de_suivi/txt'
full_documents = []
# nb_docs = len(os.listdir(path))
nb_docs=10
for file, i in zip(os.listdir(path), range(0,nb_docs)): 
    
    with open(f"{path}/{file}") as f:
        raw_text = f.read()
    full_documents.append(raw_text)
```

```python
%pip uninstall --y faiss-cpu & pip install faiss-gpu
```

```python
RAG.index(
    collection=full_documents,
    index_name="ASN",
    max_document_length=180,
    split_documents=True,
    )

```
